```python
X = np.array(train3)
y = np.array(target3['Global_Sales'])
```

```python
import lightgbm as lgbm
random_state = 42
clf = lgbm.LGBMClassifier(random_state=random_state, silent=True, metric='None', n_jobs=4)
```

```python
from sklearn.model_selection import ShuffleSplit
import matplotlib.pyplot as plt
    
ss = ShuffleSplit(n_splits = 10,
                  train_size = 0.6,
                  test_size = 0.4,
                  random_state = 0)

for train_index, test_index in ss.split(X):

    X_train, X_test = X[train_index], X[test_index]
    Y_train, Y_test = y[train_index], y[test_index]

    clf.fit(X_train, Y_train) 
```

```python
y_pred = clf.predict(np.array(test3))
y_pred = y_pred.astype(np.int)
```

```python
sub = sample_submission
sub['Global_Sales'] = list(map(int, y_pred))
sub.to_csv("lgbm7_submission.csv", index=False)
```
